```python
from bs4 import BeautifulSoup
```

```python
import os
import sys
import urllib.request
import datetime
import time
import json
import pandas as pd
```

```python
def album_crawling(result):
    for page in range(1,50): # 1p~50p 까지
        bugs_url = 'https://music.bugs.co.kr/newest/track/totalpicked?page=%d'%page
        print(bugs_url)
        html = urllib.request.urlopen(bugs_url)
        soupbugs= BeautifulSoup(html, 'html.parser')
        tag_tbody = soupbugs.find('tbody')
        for album in tag_tbody.find_all('tr'):
            if album.find('p').attrs['adult_yn'] == 'Y': #19금 노래 제거
                continue
            else:
                result.append(list(album.attrs.values()))
    return 

            
```

```python
result = []
album_crawling(result)
frame = pd.DataFrame(result, columns =('albumid','artistid' ,'mvid' ,'trackid','multiartist','rowtype'))
frame.to_csv('/home/bugs.csv', encoding = 'cp949', mode ='w', index = False, header = False)
print('finish')
```
