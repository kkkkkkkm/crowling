```python
from selenium import webdriver
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd 
import time
```

```python
wd = webdriver.Chrome('C:\\Users\\USER\\Desktop\\chromedriver.exe')
```

```python
wd.get('https://www.melon.com/chart/index.htm') #멜론 차트 Top100
time.sleep(1)
```

```python
html = wd.page_source
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
v_result = []

for i in soup.select("tbody > tr > td > div > input.input_check "):
    v_result.append(i.attrs['value'])
    
print(v_result)
```

```python
art_name = []
album_genre = []
album_flac = []
album_date = []
t_result = []

for i in soup.select("tbody > tr > td > div > a.image_typeAll"):
    t_result.append(i.attrs['title'])


for i in soup.select("tbody > tr > td > div > input.input_check "):
    a = i.attrs['value']
    wd.execute_script('melon.link.goSongDetail(%s)' %a)
    time.sleep(1)
    html2 = wd.page_source
    soup2 = BeautifulSoup(html2, 'html.parser')
    art_name.append(soup2.select("div.entry > div.info > div.artist > a")[0].attrs['title'])
    album_meta = soup2.select('div.meta > dl.list> dd ')
    album_date.append(album_meta[1].string)
    album_genre.append(album_meta[2].string)
    try:
        album_flac.append(album_meta[3].string)     #Flac가 없는 곡이 존재, 예외처리
    except IndexError:
        album_flac.append('0')
        
    
print(art_name)
print(album_genre)
print(album_flac)
print(album_date)
```

```python
df = pd.DataFrame(zip(v_result,t_result, art_name, album_genre, album_date, album_flac), 
                columns = ['곡 번호','앨범', '가수', '장르', '발매일', 'Flac'])
```
